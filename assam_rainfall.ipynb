```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:/Users/sanika khanduja/Downloads/districtwiseassamrainfall_updated.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Basic information about the dataset
print("\nBasic information:")
print(df.info())

# Summary statistics of numerical columns
print("\nSummary statistics:")
print(df.describe())

# Checking for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Distribution of 'Actual' rainfall values
plt.figure(figsize=(10, 6))
sns.histplot(df['Actual'], kde=True)
plt.title('Distribution of Actual Rainfall')
plt.xlabel('Actual Rainfall')
plt.ylabel('Frequency')
plt.show()

# Distribution of 'Normal' rainfall values
plt.figure(figsize=(10, 6))
sns.histplot(df['Normal'], kde=True)
plt.title('Distribution of Normal Rainfall')
plt.xlabel('Normal Rainfall')
plt.ylabel('Frequency')
plt.show()

# Scatter plot to see the relationship between 'Actual' and 'Normal'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Normal', y='Actual', data=df)
plt.title('Actual vs Normal Rainfall')
plt.xlabel('Normal Rainfall')
plt.ylabel('Actual Rainfall')
plt.show()

# Boxplot to see the distribution of 'Actual' rainfall across different years
plt.figure(figsize=(12, 8))
sns.boxplot(x='Year', y='Actual', data=df)
plt.title('Actual Rainfall Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Actual Rainfall')
plt.xticks(rotation=90)
plt.show()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Actual', 'Normal', 'Dep']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Top 10 districts with the highest 'Actual' rainfall
top_10_districts = df.groupby('DISTRICT NAME')['Actual'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_districts.values, y=top_10_districts.index)
plt.title('Top 10 Districts with Highest Actual Rainfall')
plt.xlabel('Average Actual Rainfall')
plt.ylabel('District Name')
plt.show()
```

```python
import seaborn as sns

# Joint plot for Actual vs Normal rainfall
sns.jointplot(x='Normal', y='Actual', data=df, kind='reg', height=8)
plt.suptitle('Joint Plot of Actual vs Normal Rainfall', y=1.02)
plt.xlabel('Normal Rainfall')
plt.ylabel('Actual Rainfall')
plt.show()
```

```python
# Calculate mean deviation for each district
district_flood_risk = df.groupby('DISTRICT NAME')['Dep'].mean().sort_values(ascending=False)

# Identify top flood-prone districts
flood_prone_districts = district_flood_risk.head(10)
print("Top 10 flood-prone districts based on deviation:")
print(flood_prone_districts)

# Bar plot of top flood-prone districts
plt.figure(figsize=(12, 8))
sns.barplot(x=flood_prone_districts.values, y=flood_prone_districts.index, palette="Reds_d")
plt.title('Top 10 Flood-Prone Districts (Based on Rainfall Deviation)')
plt.xlabel('Average Deviation from Normal Rainfall')
plt.ylabel('District Name')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for multiple years
data = {
    'District': ['BAKSA'] * 5 + ['CHIRANG'] * 5 + ['BARPETA'] * 5,
    'Year': [2019, 2020, 2021, 2022, 2023] * 3,
    'Actual Rainfall (cm)': [
        1824.5, 2326.1, 1338.5, 1716.3, 1792.3,  # BAKSA
        2682.9, 4326.5, 2257.6, 2821.2, 1902,    # CHIRANG
        2276.7, 3301.6, 1964.2, 2181.7, 2328.4,  # BARPETA
        
    ],
    'Normal Rainfall (cm)': [
        1532.8, 1532.8, 1532.8, 1532.8, 1532.8,  # BAKSA
        2425, 2425, 2425, 2425, 2425,          # CHIRANG
        2139.6, 2139.6, 2139.6, 2139.6, 2139.6,  # BARPETA
        
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Prepare data for plotting
years = df['Year'].unique()
districts = df['District'].unique()
bar_width = 0.15  # Width of each bar
index = range(len(years))

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

for i, district in enumerate(districts):
    district_data = df[df['District'] == district]
    x = [idx + i * bar_width for idx in index]  # Position of bars for each district
    ax.bar(x, district_data['Actual Rainfall (cm)'], width=bar_width, label=f'{district} - Actual')
    ax.bar([p + bar_width for p in x], district_data['Normal Rainfall (cm)'], width=bar_width, label=f'{district} - Normal')

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Rainfall (cm)')
ax.set_title('Yearwise Actual vs Normal Rainfall by District')
ax.set_xticks([idx + (len(districts) - 1) * bar_width / 2 for idx in index])
ax.set_xticklabels(years)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
